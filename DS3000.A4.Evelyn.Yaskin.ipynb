## 2. Inspect the Data (5 pts) 
Inspect the data and perform the following:

```python
import pandas as pd

tripdata_df = pd.read_parquet(path = 'green_tripdata_2023-02.parquet', 
                      engine = 'fastparquet')

print(tripdata_df.shape)

tripdata_df['VendorID'] = tripdata_df['VendorID'].astype('Int64')

tripdata_df['RatecodeID'] = tripdata_df['RatecodeID'].astype('Int64')

tripdata_df['store_and_fwd_flag'] = tripdata_df['store_and_fwd_flag'].astype('string')

tripdata_df['payment_type'] = tripdata_df['payment_type'].astype('Int64')

tripdata_df['trip_type'] = tripdata_df['trip_type'].astype('Int64')

tripdata_df['passenger_count'] = tripdata_df['passenger_count'].astype('Int64')

tripdata_df['PULocationID'] = tripdata_df['PULocationID'].astype('Int64')
tripdata_df['DOLocationID'] = tripdata_df['DOLocationID'].astype('Int64')

print(tripdata_df.dtypes)

tripdata_df.head()
```

## 4. Missing Values (10 pts) 
Evaluate the frequency of missing values in the dataset and perform the following:
1. explain if missing values are an isolated case or a widespread issue? 
2. what method would you recommend to handle missing data?
3. write the suitable code to handle the missing values.

Missing values are a widespread issue in this code since 7 columns all have a high count of missing values. 
Also, ehail_fee in particular has an 100 percent count of missing values. 

The method that I chose to handle missing data in this particular set was to calculate the percentages/values of all the missing values, create a threshold to choose the percentage of missing rows for each column to drop; then consequently, drop those rows and reprint the cleaned data. 


```python
import pandas as pd

missing_values = tripdata_df.isnull().sum()
print("Missing values:")
print(missing_values)
```

```python
missing_percent = (missing_values / len(tripdata_df)) * 100

print("Percent of missing values:")
print(missing_percent)
```

```python
# Created a threshold to delete rows with 10% or more of missing values. 

threshold = 10 # I made this 10%, because with a threshold of anything lower, certain columns
#are erased that I want to use for later questions.  

columns_to_drop = missing_percent[missing_percent > threshold].index

cleaned_df = tripdata_df.drop(columns=columns_to_drop)

print("DataFrame after handling missing values:")
print(cleaned_df)
```

# 5. Invalid Data (10 pts) 
Explore the data to determine if there are any invalid values and perform the following: 
1. demonstrate at least three issues that you found.
2. write suitable code to handle the invalid data that you found.

1. I noticed that the minimum count for the passenger_count was 0, this was weird since taxis shouldn't be driving  cars without passengers in them. 
2. I changed negative values to be positive instead, it didn't make sense for values like the total amount, tip amount, and mta tax values to be negative.
3. I also changed the trip_distance to be greater than 0 too, because it also doesn't make sense for a trip to have 0 miles driven.

```python
filtered_data = cleaned_df[
    (cleaned_df['fare_amount'] > 0) &
    (cleaned_df['extra'] > 0) &
    (cleaned_df['tip_amount'] > 0) &
    (cleaned_df['tolls_amount'] > 0) &
    (cleaned_df['mta_tax'] > 0) &
    (cleaned_df['improvement_surcharge'] > 0) &
    (cleaned_df['total_amount'] > 0) &
    # Passenger count should not be 0.
    (cleaned_df['passenger_count'] > 0) &
    # Trip distance should not be 0.
    (cleaned_df['trip_distance'] > 0)
]

duplicates = cleaned_df[cleaned_df.duplicated()]
print("Number of duplicated rows:", len(duplicates))

filtered_data.describe()
```

```python
import matplotlib.pyplot as plt

# Most common way New Yorkers request/hail a cab

# I am using my cleaned data (before filtering it from invalid data), so for
# visualization purposes, I wanted to use the data before any rows/columns were dropped. 

# This chart shows a breakdown of the most common ways that new yorkers choose to request a cab. 
# According to the graph, we can see that 1-street-hail is the most common with New Yorkers, with a count above 40,000.
# After the data had been cleaned, trip types with the value of 2 had either been invalid or missing, so it doesn't show up on the graph. 
top_pickup_locations = cleaned_df['trip_type'].value_counts()
plt.figure(figsize=(2, 2))
top_pickup_locations.plot(kind='bar', color='pink')
plt.title('Trip types')
plt.xlabel('Trip types')
plt.ylabel('# of Trips')
plt.show()

# Most common pick-up locations (top 5)
# This chart shows a breakdown of the most common pick-up location and their corresponding ID numbers. 
# It only shows the top 5 since there are too many pick-up location IDs to display. 
# According to the graph, we can see that 74, 75, 166, 41, and 95 are the most common, with #74 in the lead. 
top_pickup_locations = cleaned_df['PULocationID'].value_counts().head(5)
plt.figure(figsize=(2, 2))
top_pickup_locations.plot(kind='bar', color='skyblue')
plt.title('Most Common Pick-up Locations')
plt.xlabel('Pick-up IDs')
plt.ylabel('# of Trips')
plt.show()

# Most common drop-off locations (top 5)
# This chart shows a breakdown of the most common drop-off location and their corresponding ID numbers. 
# It also only shows the top 5 since there are too many drop-off location IDs to display. 
# According to the graph, we can see that 74, 75, 236, 41, 42 are the most common, with #74 in the lead. 
# From the PU and DO IDs we can tell that #74 is a popular location. 
top_pickup_locations = cleaned_df['DOLocationID'].value_counts().head(5)
plt.figure(figsize=(2, 2))
top_pickup_locations.plot(kind='bar', color='purple')
plt.title('Most Common Drop-off Locations')
plt.xlabel('Drop-off IDs')
plt.ylabel('# of Trips')
plt.show()
```
